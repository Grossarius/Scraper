```python
import pandas as pd
import json
import ast

# Done: "bakery items", "dairy-eggs-fridge", "deli-chilled-meals", "drinks", "freezer", "fruit-veg", "health-wellness health-foods", "lunch-box", "pantry 1", "pantry 2",
all_item = [ "poultry-meat-seafood"]

filename = "Data\Woolies Item\Woolies {}.csv"

for item in all_item:
    df = pd.read_csv(filename.format(item), header=None)
    df.columns = ["ID", "Name", "Data"]
    # Initialize an empty list to store the extracted product information
    products = []
    count = 0
    # Iterate over each row in the DataFrame
    for index, row in df.iterrows():
        print(count)
        print(item)
        count +=1
        try:
            product_data = ast.literal_eval(row["Data"])
            product = {}

            # Extract information from the 'Product' category
            product_info = product_data.get('Product', {})
            if product_info is None:
                product_info = {}
            product['Product Name'] = product_info.get('Name', "None")
            product['Display Name'] = product_info.get('DisplayName', "None")
            product['Stockcode'] = product_info.get('Stockcode', "None")
            product['Barcode'] = product_info.get('Barcode', "None")
            product['Medium Image File'] = product_info.get('MediumImageFile', "None")
            product['Cup Measure'] = product_info.get('CupMeasure', "None")
            product['Price'] = product_info.get('Price', "None")
            product['Cup Price'] = product_info.get('CupPrice', "None")
            product['Cup String'] = product_info.get('CupString', "None")
            product['Has Cup Price'] = product_info.get('HasCupPrice', "None")
            product['Package Size'] = product_info.get('PackageSize', "None")

            # Extract information from the 'SapCategories' category
            sap_categories = product_info.get('SapCategories', {})
            if sap_categories is None:
                sap_categories = {}
            product['Sap Department Name'] = sap_categories.get('SapDepartmentName', "None")
            product['Sap Category Name'] = sap_categories.get('SapCategoryName', "None")
            product['Sap Sub Category Name'] = sap_categories.get('SapSubCategoryName', "None")
            product['Sap Segment Name'] = sap_categories.get('SapSegmentName', "None")

            # Extract information from the 'PrimaryCategory' category
            primary_category = product_data.get('PrimaryCategory', {})
            if primary_category is None:
                primary_category = {}
            product['Department'] = primary_category.get('Department', "None")
            product['Aisle'] = primary_category.get('Aisle', "None")

            # Extract information from the 'AdditionalAttributes' category
            additional_attributes = product_data.get('AdditionalAttributes', {})
            if additional_attributes is None:
                additional_attributes = {}
            product['Description'] = additional_attributes.get('description', "None")
            product['Vegetarian'] = additional_attributes.get('vegetarian', "None")
            product['Wool Dietary Claim'] = additional_attributes.get('wool_dietaryclaim', "None")
            product['Allergen Contains'] = additional_attributes.get('allergencontains', "None")
            product['Allergy Statement'] = additional_attributes.get('allergystatement', "None")
            product['Ingredients'] = additional_attributes.get('ingredients', "None")

            # Append the product information to the list
            products.append(product)
        except (ValueError, SyntaxError) as e:
            print(e)
            print(row["Data"])
            pass


    # Create a DataFrame from the extracted product information
    extracted_df = pd.DataFrame(products)

    # Define the headers
    headers = ['Product Name', 'Display Name', 'Stockcode', 'Barcode', 'Medium Image File', 'Cup Measure',
            'Price', 'Cup Price', 'Cup String', 'Has Cup Price', 'Package Size',
            'Sap Department Name', 'Sap Category Name', 'Sap Sub Category Name', 'Sap Segment Name',
            'Department', 'Aisle', 'Description', 'Vegetarian', 'Wool Dietary Claim', 'Allergen Contains',
            'Allergy Statement', 'Ingredients']

    # Append the extracted DataFrame to an existing Excel file or create a new file
    excel_file = 'Woolies {} info.xlsx'  # Replace with the actual file path

    sheet_name = 'Sheet1'  # Replace with the desired sheet name
    extracted_df.to_excel(excel_file.format(item), sheet_name=sheet_name, index=False, header=headers)
    print(item)
```
